```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Upload csv file
from google.colab import files
files.upload()
```

```python
#Store data into object df
df = pd.read_csv('MyPortfolio - Trading.csv', skiprows=4, names=['Company_Name','Ticker','Shares_Amount','Purchase_Market','OG_Price','Change_USD','Change_Perc','Closing_Price','Day_Gain','Day_Gain_PERC','Total_Gain_USD','Total_Gain_Perc','Value_USD','Invest_Platform'])
df = df.style.set_properties(**{
    'background-color': 'honeydew',
    'font-size': '12pt',
})


#Display data
df
```

```python
#Get the total invested amount in both portfolios: E*Trade & Robinhood
Portfolio_Total_Amount = sum(df['Value_USD'])
Portfolio_Total_Amount = round(Portfolio_Total_Amount , 2)
Portfolio_Total_Amount
```

```python
#Visualize my portfolios/investments with a Pie Chart
stock_tickers = df['Ticker'].values
sizes = df['Value_USD']

listOfZeros = [0] * df.shape[0]
n = random.randint(0, df.shape[0]-1)
listOfZeros[n] = 0.1
explode = listOfZeros

#Create the pie chart
fig1 , ax1 = plt.subplots(figsize = (15,15))

#Plot the pie chart
ax1.pie(sizes, explode=explode, labels= stock_tickers, autopct = '%.2f%%', shadow='True', startangle=360)

#Set the Title
ax1.set_title('My Portfolio', color='Green', fontsize= 22)

#Overview
x= -1.75
y= 1
ax1.text(x,y, 'Overview:', fontsize=24, color='Green')

y_counter = 0.12
ax1.text(x,y-y_counter, 'Total: $'+str(Portfolio_Total_Amount), fontsize=15, color='Teal')


for i in range(0, df.shape[0]):
  ax1.text(x, 0.93 - y_counter, df['Company_Name'][i] + ': $' + str(df['Value_USD'][i]), fontsize=12, color='Black') 
  y_counter = y_counter + 0.05
```
