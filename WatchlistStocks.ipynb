```python
#Plot data from CSV file into bar graph
data = pd.read_csv('Trading - Watchlist1.csv')
df = pd.DataFrame(data)

#This defines what each axis will be. X-axis: column 1[Ticker]. Y-axis: column 4 [Projected_Growth_FiveYears]
X = list(df.iloc[:, 1])  
Y = list(df.iloc[:, 4])
  
# Plot the data using bar() method
plt.figure(figsize=(28,13))

#Change the color of the bars
plt.bar(X, Y, color='teal')

#Add title and labels
plt.title("Watchlist & Projections", fontsize=20, fontweight='bold')
plt.xlabel("Ticker", fontsize=15)
plt.ylabel("Projected Growth - 5 yrs ($)", fontsize=15)

#Rotate the 'ticker' labels at a 45 degree angle for better legibility
plt.xticks(rotation=45)
sns.set()

#Change the background color
ax = plt.axes()
ax.set_facecolor('gainsboro')

# Display the plot
plt.show()
```
